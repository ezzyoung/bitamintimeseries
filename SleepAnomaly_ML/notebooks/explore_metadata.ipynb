# 라벨링

0 - 정상
1 - 이상

```python
# 먼저 Diagnosis를 결측치 채우기
df_subgroup1['Diagnosis'] = df_subgroup1['Diagnosis'].fillna('Healthy')
df_subgroup2['Diagnosis'] = df_subgroup2['Diagnosis'].fillna('Healthy')
df_subgroup3['Diagnosis'] = df_subgroup3['Diagnosis'].fillna('Healthy')
```

```python
# 라벨링 : Healthy면 0, 나머지는 1
df_subgroup1['has_disorder'] = df_subgroup1['Diagnosis'].apply(lambda x : 0 if str(x).strip().lower() == 'healthy' else 1)
df_subgroup2['has_disorder'] = df_subgroup2['Diagnosis'].apply(lambda x : 0 if str(x).strip().lower() == 'healthy' else 1)
df_subgroup3['has_disorder'] = df_subgroup3['Diagnosis'].apply(lambda x : 0 if str(x).strip().lower() == 'healthy' else 1)
```

### 전처리 메타데이터 새로 저장

```python
df_subgroup1.to_csv('../Database/processed/meta/Details_subgroup1.csv')
df_subgroup2.to_csv('../Database/processed/meta/Details_subgroup2.csv')
df_subgroup3.to_csv('../Database/processed/meta/Details_subgroup3.csv')
```
